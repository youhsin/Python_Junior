```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def L(x):
    return x*(x-1)*(x-2)*(x-3)*(x-4)
def dL(x):
    return 5*x**4-40*x**3+105*x**2-100*x+24
def gd(x):
    return x-(5*x**4-40*x**3+105*x**2-100*x+24)*0.001
x=np.arange(-2,5,0.01)

plt.plot(x,L(x))
plt.ylim(-20,20)
plt.grid()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
def gradient_descent(x,lr,epoch):
    x_gd=[]
    y_gd=[]      ##建立空列表來收集結果
    
    x_gd.append(x)
    y_gd.append(L(x))  ##收集第一次的結果
    
    for i in range(epoch):
        x=x-lr*dL(x)
        x_gd.append(x)   ##收集每次迭代的結果
        y_gd.append(L(x))
        
    return x_gd,y_gd
```

```python
def draw(x0,lr,epoch):
    x=np.arange(-2,5,0.01)
    
    plt.ylim(-5,5)
    plt.xlim(-1,5)
    plt.grid()
    plt.xlabel("$x$")
    plt.ylabel("$y$")
    
    
    x_gd, y_gd=gradient_descent(x0,lr,epoch)    ##  x被用過改x0
    
    plt.plot(x,L(x))
    plt.scatter(x_gd,y_gd,c="r")
```

```python
draw(1,0.01,10)  #畫圖
gradient_descent(1,0.01,10)  #看值
```

```python
def optimize():    #定义最优化函数
#建空列表收集x,y,z的值
    z_list = []     
    x_list = []
   
    for x in range(-10,10,1):
        
        z = x**2
        z_list.append(z)
        x_list.append(x)
        
    return z_list,x_list

z_list,x_list = optimize()
#使用python自带的min()函数，找出z_value这个列表中最小的值，最小值为0
min_value = min(z_list)
#通过list.index(值)找出这个值在列表list中的索引值
index = z_list.index(min(z_list))
#因为x,y的索引值跟z是一样的，所以这样对应的索引值也是x,y的索引值
#因此将索引值带入得到的x,y就是使得z最小的时候的值
x_value = x_list[index]

print(min_value,index,x_value)
```
