## 利用矩陣取出相應範圍內的數字作為題目，再取數字做每一次的答案


```python
import numpy as np
from ipywidgets import interact_manual
```

輸入有幾位玩家就生成幾組題目

```python
players=int(input("玩家數>>"))
matrix=int(input("想玩幾宮格>>"))
num=int(input("最大的數字為>>"))
col_row=int(matrix**(1/2))

ans=np.random.choice(np.arange(1,num+1),size=num,replace=False)


for i in range(players):
    print(np.random.choice(np.arange(1,num+1),size=matrix,replace=False).reshape(col_row,col_row))
    print("")

for i in range(num):
    print(ans[i])
```
