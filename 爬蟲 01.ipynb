```python
import requests
from bs4 import BeautifulSoup
```

```python
r=requests.get('https://www.ptt.cc/bbs/MobileComm/index.html')#將網頁get下來
soup=BeautifulSoup(r.text,'html.parser')#將網頁資料轉換成HTML parser
sel=soup.select('div.title a')#取HTML標中的 <div class="title"></div> 中的<a>標籤存入sel
for s in sel:
    print(s["href"],s.text)
```

```python
url="https://www.ptt.cc/bbs/MobileComm/index6511.html"
for i in range(3):
    r=requests.get(url)
    soup=BeautifulSoup(r.text,"html.parser")
    sel=soup.select("div.title a")
    u=soup.select("div.btn-group.btn-group-paging a")
    print("本頁網址為"+url)
    url="https://www.ptt.cc"+u[1]["href"]
ptt=open("ptt爬.txt","w")
for s in sel:
    print(s.text)
    ptt.write(s.text)
ptt.close()
```
