# matplotlib 相關功能

```python

x=[2,3,1,4,]
y1=[1,2,3,4]
y2=[3,5,8,6]
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.figure(figsize=(5,5),facecolor="b")
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.figure(figsize=(5,5))

plt.plot(x,y1,label="y1")
plt.plot(x,y2,label="y2")

plt.legend(loc="best")
```

```python
plt.figure(figsize=(5,5))

plt.plot(x,y1,label="y1")
plt.plot(x,y2,label="y2")

plt.legend(loc="best") #可以用upper left...其他的位置
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("title")
```

```python
plt.figure(figsize=(5,5))

plt.plot(x,y1,label="y1")
plt.plot(x,y2,label="y2")

plt.legend(loc="best") #可以用upper left...其他的位置
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("title")

plt.xlim(-2,20)
plt.ylim(20,-10)  #控制XY的顯示範圍
```

```python
plt.figure(figsize=(5,5))

plt.plot(x,y1,label="y1")
plt.plot(x,y2,label="y2")

plt.legend(loc="best") #可以用upper left...其他的位置
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("title")


plt.xticks(np.arange(-2,20,5))
plt.yticks(np.arange(20,-10,5)) #控制顯示範圍還可以控制間隔
```

```python
import seaborn as sns
```

```python
sns.set_style('whitegrid') #決定圖形風格
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
tips=sns.load_dataset('tips');
tips.head #seabon的資料庫(此為一個餐廳的小費database)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter');#xy為database的資料(colunm) jointplot就是以下呈現的圖的風格
```
